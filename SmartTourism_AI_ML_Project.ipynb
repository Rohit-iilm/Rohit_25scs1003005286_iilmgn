```python
# Install libraries (if needed)
!pip install -q scikit-learn pandas matplotlib

# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor

# Step 1: Create sample tourism dataset
data = {
    "bookings_lag1": [100, 120, 130, 150, 170, 180, 200],
    "price": [2000, 2100, 1900, 2200, 2300, 2500, 2700],
    "temperature": [25, 28, 30, 27, 26, 29, 31],
    "bookings": [120, 130, 150, 170, 180, 200, 220]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)

# Step 2: Train model
X = df[["bookings_lag1", "price", "temperature"]]
y = df["bookings"]

model = RandomForestRegressor(n_estimators=100)
model.fit(X, y)

# Step 3: Prediction
prediction = model.predict([[190, 2600, 28]])
print("\nPredicted bookings:", prediction)

# Step 4: Visualization
plt.plot(df["bookings"], marker='o')
plt.title("Tourism Demand Trend")
plt.xlabel("Days")
plt.ylabel("Bookings")
plt.show()
```
